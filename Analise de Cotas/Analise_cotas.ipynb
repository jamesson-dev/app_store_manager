```python
import pandas as pd
import random
import warnings
```

```python
df = pd.read_excel('status/Cotas_ativas_pre_cancelados.xlsx')
```

```python
# Verificar colunas disponíveis
print("Colunas originais:", df.columns.tolist())
```

```python
# Função para gerar CPF anonimizado (ex.: Vendedor_001)
def anonimizar_cpf(cpf_series):
    unique_cpfs = cpf_series.unique()
    cpf_map = {cpf: f"Vendedor_{i+1:03d}" for i, cpf in enumerate(unique_cpfs)}
    return cpf_series.map(cpf_map)

# Função para gerar telefone anonimizado no formato (XX) 9XXXX-XXXX
def anonimizar_telefone(telefone_series):
    def gerar_telefone():
        ddd = random.choice(["00"])
        numero = f"9{random.randint(1000, 9999)}-{random.randint(1000, 9999)}"
        return f"({ddd}) {numero}"
    
    unique_telefones = telefone_series.dropna().unique()
    telefone_map = {tel: gerar_telefone() for tel in unique_telefones}
    return telefone_series.map(telefone_map).fillna(telefone_series)

# Anonimizar os campos
if 'CPF Vendedor' in df.columns:
    df['CPF Vendedor'] = anonimizar_cpf(df['CPF Vendedor'])

if 'Telefone' in df.columns:
    df['Telefone'] = anonimizar_telefone(df['Telefone'])

if 'Telefone1' in df.columns:
    df['Telefone.1'] = anonimizar_telefone(df['Telefone.1'])

if 'Celular' in df.columns:
    df['Celular'] = anonimizar_telefone(df['Celular'])
    
# Verificar colunas disponíveis
print("Colunas originais:", df.columns.tolist())

# Salvando o arquivo separado com os dados anonimizados
df.to_excel("Cotas_ativas_pre_cancelados_Anonimizado.xlsx", index=False)
print("\nArquivo anonimizado salvo como 'Cotas_ativas_pre_cancelados_Anonimizado.xlsx'")
```
